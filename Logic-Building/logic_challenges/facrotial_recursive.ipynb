### Factorial 

fact(4) = 4 * fact(3)
        = 4 * 3 * fact(2)
        = 4 * 3 * 2 * fact(1)
        = 4 * 3 * 2 * 1 * fact(0)
        = 4 * 3 * 2 * 1 * 1
        = 2

```python
def fact (n):
    if n ==  0 :
        return  1
    return  n * fact(n-1)

nums = int(input("enter  the number:"))
fact(nums)
```
