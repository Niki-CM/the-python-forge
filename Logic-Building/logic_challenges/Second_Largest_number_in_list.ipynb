### Second Largest Number in the List


```python
l= [6,4,3,2,2,34,5,3,2,1,34,3,123,43,221]
l.sort()
print(f"Second Largest number:{l[-2]}")
```

### Another way


```python
nums =[3,4,5,1,3,4,9,7,10,2,5,6,3,5,7,4,6,5,1]
nums.sort()
nums[-2]
```

### Another way

```python
nums =[3,4,5,1,3,4,9,7,10,2,5,6,3,5,7,4,6,5,1]
print(sorted(nums)[-2])
```
