### print whether the number is even or odd along with positive/negative/zero!

```python
num = int(input("enter the number"))
if num  % 2 == 0 :
    if num > 0 :
        print(f"{num} is Even Positive Number")
    elif num < 0 :
        print(f"{num} is Even Negative  Number")
    else:
        print(f"{num} It's a Zero")
else:
    if num > 0 :
        print(f"{num} is Odd Positive Number")
    elif num < 0 :
        print(f"{num} is Odd Negative  Number")
    else:
        print(f"{num} It's a Zero")
        
```
