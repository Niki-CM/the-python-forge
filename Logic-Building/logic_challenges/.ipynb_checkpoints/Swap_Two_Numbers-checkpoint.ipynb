### Swap Two Numbers

```python
a = 10
b = 14
a,b = b,a
print(f"a: {a}")
print(f"b: {b}")
```

### Second method
Using a temp variable

```python
a = 9
b = 10
temp = a
a = b
b = temp
print(a,b)
```

### Using the airthemtic method


```python
a = 9
b = 10
a = a + b
b = a - b
a = a - b
print(a,b)
```

### Try swapping 3 variables
Take 3 variables x=5, y=10, z=15 and rotate them → x=y, y=z, z=x in

```python
x = 5
y = 10
z = 15
x,y,z = y,z,x
print(x,y,z)
```
